```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
df=pd.read_csv('/home/pc480/Downloads/iris.csv')
df.columns=['sepal_length','sepal_width','petal_length','petal_width','species']
```

```python
x=df[['sepal_length','sepal_width','petal_length','petal_width']]
y=df['species']
```

```python
pca=PCA(n_components=3)
x_r=pca.fit_transform(x)
print("explained varience ratio to see contribution ratio:",pca.explained_variance_ratio_)
```

```python
colors={'Iris-setosa':'navy','Iris-veriscolor':'turquoise','Iris-virginica':'darkorange'}
```

```python
plt.figure(figsize=(8,6))
for species in y.unique():
    idx=y==species
    plt.scatter(x_r[idx,0],x_r[idx,1],c=colors.get(species),label=species,edgecolor='k',s=50)
plt.ylabel('Principal Component 2')
plt.xlabel('Principal Component 1')
plt.title('PCA of IRIS dataset')
plt.legend(loc='best')
plt.show()
plt.savefig('./')
plt.close()
```
